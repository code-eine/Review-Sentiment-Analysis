```python
#import all the necessary packages
import pandas as pd
import numpy as np
from textblob import TextBlob
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#read data from a CSV file
data = pd.read_csv(r'C:\Files\Mpesa App Reviews.csv')
data = data.head(1500)
data.dropna()
```

```python
#show the first 20 reviews
data.head(35)
```

```python
#review score plot
sns.displot(data['score'],)
```

```python
#review score count plot
sns.countplot(x='score', data=data)
```

```python
#specific review check
data['content'][15]
```

```python
#sentiment analysis on a specific review
text_blob_object = TextBlob(data['content'][15])
print(text_blob_object.sentiment)
```

```python
#add a sentiment polarity column to the review data
def find_pol(review):
    return TextBlob(review).sentiment.polarity
data ['Sentiment_Polarity'] = data['content'].apply(find_pol)
data.head()
```

```python
#review sentiment polarity count
sns.displot(data['Sentiment_Polarity'],)
```

```python
#barplot of the review sentiment polarity score
sns.barplot(x='score', y='Sentiment_Polarity', data=data)
```

```python
#find the 5 most negative reviews
most_negative = data[data.Sentiment_Polarity == -0.5].content.head()
print(most_negative)
```
